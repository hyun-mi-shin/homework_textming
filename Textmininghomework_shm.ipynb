```python
from collections import Counter

NUM = 2
BRANCHES = ["서울신문", "문화일보", "내일신문", "매일경제"]

for branch in BRANCHES:
    wordsets = academics["publisher"][branch_idxes[branch]]
    ngram_count = Counter()

    for wordset in wordsets:
        ngrams = list(zip(*[wordset[i:] for i in range(NUM)]))
        ngram_count.update(ngrams)

    print("========= {} =========".format(branch))
    for (word, count) in ngram_count.most_common(20):
        print("{}\t\t{}".format(" & ".join(word), count))
    print()
    print()
```

```python
#Pointwise Mutual Information

import math

wordsets = {}

for branch in BRANCHES:
    wordsets[branch] = academics["publisher"][branch_idxes[branch]]

    unigram_count = Counter()
    bigram_count = Counter()

    ppmis = Counter()

    for wordset in wordsets[branch]:
        unigram_count.update(wordset)

        bigram = list(zip(*[wordset[i:] for i in range(2)]))
        bigram_count.update(bigram)

    N = sum(unigram_count.values())

    for (left_w, right_w), count in bigram_count.items():
        pmi = math.log2((N * count) / (unigram_count[left_w] * unigram_count[right_w]))
        ppmi = max(pmi, 0.0)
        ppmis[(left_w, right_w)] = ppmi

    print("========= {} =========".format(branch))
    for (bigram, count) in ppmis.most_common(10):
        print("{}\t\t{}".format(" & ".join(bigram), count))
    print()
    print()
```

```python
from pprint import pprint
from gensim.models.phrases import Phrases
from gensim.models.phrases import Phraser

for branch in BRANCHES:
    gen_phrases = Phrases(wordsets[branch], min_count=5, threshold=0.4, scoring="npmi") # npmi 양수로 가면 association이 강함.
    gen_bigram = Phraser(gen_phrases) # Training된 것으로 적용

    print("================ {} ================".format(branch))
    for bi_sent, sent in zip(gen_bigram[wordsets[branch]][:10], wordsets[branch][:10]):
        if len(sent) > 2:
            print("unigram: {}".format(sent))
            print("bigram: {}".format(bi_sent))
            print()
    print("=====================================")
    print()
```

```python
from collections import Counter
from collections import defaultdict

from itertools import combinations

from operator import itemgetter

cooccurance_count = {}
cooccurance_list = {}

for branch in BRANCHES:

    cooc_count = defaultdict(Counter)

    for wordset in wordsets[branch]:
        wordset = sorted(set(wordset))

        for word_l, word_r in combinations(wordset, 2):
            cooc_count[word_l][word_r] += 1

    cooccurance_count[branch] = cooc_count

    cooc_list = []
    for word_l, word_r_count in cooccurance_count[branch].items():
        for word_r, count in word_r_count.items():
            cooc_list.append((word_l, word_r, count))

    cooccurance_list[branch] = sorted(cooc_list, key=itemgetter(2), reverse=True)
```

```python
for branch in BRANCHES:
    print("================ {} ================".format(branch))

    for (word_l, word_r, count) in cooccurance_list[branch][:10]:
        print("{} \t+\t {} \t=\t {}".format(word_l, word_r, count))

    print("=====================================")
    print()
```

```python
CHK_WORD = "관리"
for branch in BRANCHES:
    print("================ {} ================".format(branch))

    for (word_l, word_r, count) in cooccurance_list[branch][:50]:
        if (CHK_WORD in word_l) or (CHK_WORD in word_r):
            print("{} \t+\t {} \t\t=\t\t {}".format(word_l, word_r, count))

    print("=====================================")
    print()
```

```python
#Network Analysis
#네트워크 생성 with NetworkX

import sys
import math
import networkx as nx
import matplotlib.pyplot as plt

# warnings 안보이게 해줌
import warnings
import matplotlib.cbook
warnings.filterwarnings("ignore", category=matplotlib.cbook.mplDeprecation)

NODE_NUM = 80

general_networks = {}
for branch in BRANCHES:
    print("================== {} ==================".format(branch))
    G = nx.Graph()

    for word_l, word_r, count in cooccurance_list[branch][:NODE_NUM]:
        G.add_edge(word_l, word_r, weight=count)

    general_networks[branch] = G

    sizes = [math.log(G.degree(n) + 1) * 2500 for n in G.nodes()]
    plt.figure(figsize = (20, 20))

    nx.draw_networkx(G,
                    pos = nx.kamada_kawai_layout(G, scale=1),
                    node_size = sizes,
                     node_color = '#e0e0e0',
                     edge_color = 'grey',
                     font_family = "NanumGothic", # "malgun gothic" 또는 "AppleGothic"
                     #label_pos = 0.5, # [0 - 1], Head to Tail
                     with_labels = True,
                     font_size = 18
                    )

    plt.axis("off")
    plt.show()
```
